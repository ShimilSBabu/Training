# **Implementation of multiple linear regression**


# Imports

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
```

# Variables Initialization

```python
TRUE_w_1 = 3.0
TRUE_w_2 = 14.0
TRUE_b = 1.5
EXAMPLE_COUNT = 1000
x_1s = tf.random.normal(shape = [EXAMPLE_COUNT])
x_2s = tf.random.normal(shape = [EXAMPLE_COUNT])
ys = TRUE_b + TRUE_w_1 * x_1s + TRUE_w_2 * x_2s
```

# Class definition

```python
class Model(object):
  def __init__(self):
    self.weight_1 = tf.Variable(random.random())
    self.weight_2 = tf.Variable(random.random())
    self.bias = tf.Variable(random.random())

  def __call__(self, x1, x2):
    return self.bias + self.weight_1 * x1 + self.weight_2 * x2
```

# Functions

```python
def loss(predicted_y, target_y):
  return tf.reduce_mean(tf.square(predicted_y - target_y))
```

```python
def plot_data(input_1, input_2, outputs, predicted_outputs):
    real = plt.scatter(input_1, input_2, outputs, c = 'b')
    predicted = plt.scatter(input_1, input_2, predicted_outputs, c = 'r')
    plt.legend((real, predicted), ('Real Data', 'Predicted Data'))
    plt.show()
```

```python
def train(model, input_1, input_2, outputs, learning_rate):
    with tf.GradientTape() as t:
      current_loss = loss(model(input_1, input_2), outputs)
    dw_1, dw_2, db = t.gradient(current_loss, [model.weight_1, model.weight_2, model.bias])
    model.weight_1.assign_sub(learning_rate * dw_1)
    model.weight_2.assign_sub(learning_rate * dw_2)
    model.bias.assign_sub(learning_rate * db)

    return current_loss
```

# Instantiation

```python
model = Model()
```

# Plotting the initial states

```python
plot_data(x_1s, x_2s, ys, model(x_1s, x_2s))
current_loss = loss(model(x_1s, x_2s), ys).numpy()
print(f'Current Loss is : {current_loss}')
```

# Training

```python
list_weight_1, list_weight_2, list_bias = [], [], []
epochs = 20
losses = []
learning_rate = 0.1

for epoch in range(epochs):
  list_weight_1.append(model.weight_1.numpy())
  list_weight_2.append(model.weight_2.numpy())
  list_bias.append(model.bias.numpy())
  current_loss = train(model, x_1s, x_2s, ys, learning_rate = learning_rate)
  losses.append(current_loss)
  plot_data(x_1s, x_2s, ys, model(x_1s, x_2s))
  print(f'Epoch : {epoch} Weight_1 : {list_weight_1[-1]} Weight_2 : {list_weight_2[-1]} Bias : {list_bias[-1]} Current Loss : {current_loss} \n\n\n')
```

# Plotting the final states

```python
plot_data(x_1s, x_2s, ys, model(x_1s, x_2s))
current_loss = loss(model(x_1s, x_2s), ys).numpy()
print(f'Current Loss is : {current_loss}')
```

```python
plt.plot(epochs, list_weight_1, 'r', 
         epochs, list_weight_2, 'g', 
         epochs, list_bias, 'b', )
plt.plot(TRUE_w_1 * len(epochs), 'r--', 
         TRUE_w_2 * len(epochs), 'g--', 
         TRUE_b * len(epochs), 'b--', )
plt.legend(['Predicted Weight 1', 'Predicted Weight 2', 'Predicted Bias', 'True Weight 1', 'True Weight 2', 'True Bias'])
plt.show()
```
